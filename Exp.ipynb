Testing Your Code

I created a simple Python function called add_numbers inside a file named Test1.py.
This function takes two numbers and returns their sum. This is the code I chose to test for this assignment.

Writing Your First Test

I wrote a separate file called Test2.py that contains a unittest class called TestAddNumbers. Inside this class, I wrote a method named test_add which checks that the add_numbers function returns the correct value.

Since unittest does not run correctly inside Jupyter Notebook, I created and ran the test in VS Code.

Executing Your First Test

In the VS Code terminal, I ran the command:

python -m unittest Test2

This command automatically discovers my test file and runs the test. The test executed successfully and showed one passing test.

My Test Results Explained

The terminal output displayed:

.

Ran 1 test in 0.000s

OK

The dot represents one test that passed.
“OK” means there were no errors or failures.
This tells me that my function add_numbers is working correctly because it returned the expected output during the test.
